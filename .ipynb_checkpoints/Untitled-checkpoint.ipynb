```python
from torch.utils.data.dataset import Dataset
import os
import pandas as pd
import io
class MyCustomDataset(Dataset):
   
    def __init__(self):
        
        
        self.landmarks_frame = pd.read_csv(os.path.join('/home/ubuntu/SCUT5500',
                                                     'train_test_files/split_of_60%training and 40%testing/train.txt'),sep=' ', names = ['image','score'],header = None)
        #self.landmarks_frame=self.landmarks_frame.head(10)
        self.root_dir = '/home/ubuntu/SCUT5500/Images'
 
    
    def __len__(self):
        return len(self.landmarks_frame)
    
    def __getitem__(self, index):
        img_name = self.landmarks_frame.iloc[index, 0]
        image = io.imread(img_name)
        landmarks = self.landmarks_frame.iloc[index, 1:]
        sample = {'image': image, 'scores': landmarks}
        return sample
    
train_dataset = MyCustomDataset()
train_dataset[0]
```
